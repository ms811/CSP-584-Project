```python
import pandas as pd
from neo4j import GraphDatabase   
```

```python
# Creating Product Nodes

driver = GraphDatabase.driver("bolt://localhost", auth=("neo4j", "csp584"))
data = pd.read_csv("C:\\Users\\ms811\\Desktop\\groceryhub\\groceryhub.csv", encoding="ISO-8859-1")

data_category_range = data['product_name'].unique()
data_category_range = data_category_range.tolist()

for i,value in enumerate(data_category_range):
    data[data['product_name'] == value].to_csv(r'product_name_'+str(value)+r'.csv',index = False, na_rep = 'N/A')
print(data)

with driver.session() as session:
    deleteproducts = """ MATCH (n:Product) DETACH DELETE n """
    session.run(deleteproducts)
    
    cqlCreate = """ CREATE (p1:Product { Id:$Id, name:$name, productDescription:$productDescription, productCurrentPrice: $productCurrentPrice, productDiscount:$productDiscount, productActualPrice:$productActualPrice, productCategory:$productCategory, productManufacturer:$productManufacturer, productInventory:$productInventory, storeZipCode:$storeZipCode, rating:$rating}) """
    for each in range(len(data)):
        
        session.run(cqlCreate,number = each+1, Id = str(data.product_id[each]), name = str(data.product_name[each]), productDescription = str(data.product_description[each]), productCurrentPrice = str(data.product_currentprice[each]), productDiscount = str(data.product_discount[each]), productActualPrice = str(data.product_actualprice[each]), productCategory = str(data.product_category[each]), productManufacturer = str(data.product_manufacturer[each]), productInventory = str(data.inventory[each]), storeZipCode = str(data.store_zipcode[each]), rating = str(data.rating[each])) 
        
```

```python
# Creating Customer nodes

data = pd.read_csv("C:\\Users\\ms811\\Desktop\\groceryhub\\login.csv", encoding="ISO-8859-1")

data_category_range = data['user'].unique()
data_category_range = data_category_range.tolist()

for i,value in enumerate(data_category_range):
    data[data['user'] == value].to_csv(r'user_'+str(value)+r'.csv',index = False, na_rep = 'N/A')
print(data)

with driver.session() as session:
    deleteCustomers = """ MATCH (n:Customer) DETACH  DELETE n """
    session.run(deleteCustomers)
    cqlCreate = """ CREATE (c1:Customer { name: $name, zip: $zipcode, email: $email }) """
    for each in range(len(data)):
        session.run(cqlCreate, number = each+1,name = str(data.user[each]), zipcode = str(data.zipcode[each]), email = str(data.email[each])) 
```

```python
from pandas import DataFrame

print("Get the number of transactions that were NOT delivered on time for every Delivery Zip Code:")
print("")
with driver.session() as session:
    query1 = """MATCH (customer:Customer)-[transaction:BOUGHT]->(product)
                WHERE transaction.transaction_status = "Disputed"
                WITH customer, count(*) AS number_of_late_delivery
                WHERE number_of_late_delivery > 0
                WITH customer,number_of_late_delivery
                RETURN (customer.zip) AS zipcode , count(number_of_late_delivery) as number_of_late_delivery"""
    dtf_data = DataFrame([dict(_) for _ in session.run(query1)])
    print(dtf_data)
```

```python
print("Get the number of disputed transactions and the list of customer names for these disputed transactions for Delivery Zip Code: ")
print("")
with driver.session() as session:
    query2 = """MATCH (login:Customer)-[customerorders:BOUGHT]->(groceryhub)
                WHERE customerorders.transactionStatus = "Disputed"
                WITH login, groceryhub, customerorders
                RETURN login.user"""
    dtf_data = DataFrame([dict(_) for _ in session.run(query2)])
    print(dtf_data)
```

```python
print("Get the number of disputed transactions for EVERY CUSTOMER that has more that one disputed transaction: ")
print("")
with driver.session() as session:
    query3 = """MATCH (customer:Customer)-[transaction:BOUGHT]->(product)
                WHERE transaction.transaction_status = "disapproved"
                WITH customer, count(*) AS number_of_disputed_transactions
                WHERE number_of_disputed_transactions > 1
                RETURN customer.name AS Customer, number_of_disputed_transactions"""
    dtf_data = DataFrame([dict(_) for _ in session.run(query3)])
    print(dtf_data)
```

```python
print("Get the top 3 customers that reported the maximum number of returned orders")
print("")
with driver.session() as session:
    query4 = """MATCH (customer:Customer)-[transaction:BOUGHT]->(product)
                WHERE transaction.order_returned= "True"
                WITH customer, count(*) AS number_of_orders_returned
                ORDER BY number_of_orders_returned DESC LIMIT 3
                RETURN customer.name AS `Customer Name`, number_of_orders_returned"""
    dtf_data = DataFrame([dict(_) for _ in session.run(query4)])
    print(dtf_data)
```

```python
print("Get the number of returned orders and got review rating 1 for every product category in every Delivery Zip Code ")
print("")
with driver.session() as session:
    query5 = """MATCH (customer:Customer)-[transaction:BOUGHT]->(product)
                WHERE transaction.reviewrating = "1" and transaction.order_returned = "True"
                WITH customer, count(*) AS number_of_transactions, transaction
                WHERE number_of_transactions > 0
                WITH customer,number_of_transactions,transaction
                RETURN (customer.zip) AS zipcode ,count(number_of_transactions) as number_of_returned_orders ,transaction.categoryname as category"""
    dtf_data = DataFrame([dict(_) for _ in session.run(query5)])
    print(dtf_data)
```
